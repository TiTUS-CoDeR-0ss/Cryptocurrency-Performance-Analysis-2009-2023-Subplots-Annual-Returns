```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Download data
data = yf.download("AAPL", start="1980-01-01", end="2023-12-31")

# Clean data (assuming you want to remove rows with missing values)
clean_data = data.dropna()

# Group by year and calculate average yearly price
yearly_averages = clean_data.groupby(pd.Grouper(freq='Y'))['Close'].mean()

# Create a DataFrame from the yearly averages
df = pd.DataFrame(yearly_averages)
df.index.name = 'Year'  # Set the index name to 'Year'

# Plot the yearly averages
plt.figure(figsize=(10, 6))
plt.plot(df['Close'], marker='o')  # Plot the 'Close' column
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('AAPL Average Yearly Price (1980-2023)')
plt.legend()
plt.grid(True)
plt.savefig('stock_price_chart.png')  # Save the plot
plt.show()
```
